### **Drop and Add New Columns**

### **Duplicate Data**

### **Handling Missing Values**

> There are 2 ways to handle missing values:
1. Delete rows or columns that contain missing values
2. Fill in missing data

> Identify and how to handel missing values:
- **name** column has <3% missing values : we will try to fill in the empty names based on the data available in the dataset. It is possible that one **host_name** has a property with the same name.
- **host_name** column only has 1 missing value : we will fill in the missing data as **Unknown Name** in that row.
- The columns **last_review, reviews_per_month, last_review(Year) & last_review(Month)** have quite high missing values (36%), in this case deleting the columns is not recommended because it can eliminate some of the data that will be analyzed. Therefore, we will fill the rows with **NaN values with 0**.

```python
#Fill in the NaN value in the "reviews per month, last_review(Year) and last review(Month)" column with 0.
airbnb['reviews_per_month'].fillna(value=0,inplace=True)
airbnb['last_review(Month)'].fillna(value=0,inplace=True)
airbnb['last_review(Year)'].fillna(value=0,inplace=True)
```

```python
#Check if there are still missing values on "reviews_per_month,last_review(Year) & last_review(Month)" columms.
airbnb.isna().sum()
```

```python
#Fill in NaN value on "host_name" column with "Unknown_name".
airbnb['host_name'] = airbnb['host_name'].fillna('Unknown_name')
```

### **Room_Type Analysis**

```python
#The most common room type on Airbnb
airbnb_room = airbnb.groupby('room_type')['room_type'].value_counts().reset_index(name='total_rooms').sort_values('total_rooms',ascending=False)

# Calculate the total of the rooms
total_rooms = airbnb['room_type'].count()

# Create a percentage column
airbnb_room['percentage'] = round((airbnb_room['total_rooms'] / total_rooms),3) * 100
airbnb_room
```

```python
#Display graphic
#Create Bar Plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='room_type', y='percentage', data=airbnb_room, palette='pastel')

# Adds a percentage value in each bar
for i, value in enumerate(round(airbnb_room['percentage'], 3)):
    plt.text(i, value/2, f'{value:.1f}%', ha='center', va='center', fontsize=10)

#Add label
plt.title('Percentage of Room Types in Airbnb Listings')
plt.xlabel('Room Type')
plt.ylabel('Percentage')

plt.show()
```

```python
#number of reviews per room_type

# Filter data that has a 'number_of_reviews' value greater than 0
filtered_airbnb = airbnb[airbnb['number_of_reviews'] > 0]

# Grouping and calculating the number of reviews based on room type
review_roomtype = filtered_airbnb.groupby('room_type')[['number_of_reviews']].sum().sort_values('number_of_reviews', ascending=False)
review_roomtype
```

```python
#Display graphic
plt.figure(figsize=(8, 5))
explode = (0.1, 0.1, 0.2, 0.2)
labels = [f"{number_of_reviews}\n({sum})" for number_of_reviews, sum in zip(review_roomtype.index, review_roomtype['number_of_reviews'])]
sizes = review_roomtype['number_of_reviews'].values

# Define custom colors
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, shadow=True, colors=colors)
plt.title('Proportion of Number of reviews per Room Type', size=15)
plt.show()
```

- **The most common room_type** in Bangkok available on the Airbnb platform is **Entire home/apartment**.
- From the number of reviews obtained, it shows that many Airbnb users choose **Entire home/apartment** as one of the most popular places to stay when in Bangkok.
- This shows that **Entire home/apartment** is the most popular and most reviewed room type, indicating a high preference from Airbnb users visiting Bangkok.
